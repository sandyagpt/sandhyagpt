```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\GPT BANTWAL\\Downloads\\Cars93-new.csv')
columns=["Manufacturer","Model","Type","Price","MPG.city","MPG.highway","Horsepower","Rear.seat.room","Passengers"]
df[columns].head()         
```

```python
df["Price"].plot(kind="box",figsize=(10,7))
```

```python
fig,ax=plt.subplots(2, 3)
fig.set_figwidth(10)
fig.set_figheight(7)
fig.suptitle("multiple boxplots",fontsize=16)
ax[0][0].boxplot(df["Price"][df["Type"]=="Compare"])
ax[0][0].set_title('Compact')
ax[0][1].boxplot(df["Price"][df["Type"]=="Large"])
ax[0][1].set_title('Large')
ax[0][2].boxplot(df["Price"][df["Type"]=="Midsize"])
ax[0][2].set_title('Midsize')
ax[1][0].boxplot(df["Price"][df["Type"]=="Small"])
ax[1][0].set_title('Small')
ax[1][1].boxplot(df["Price"][df["Type"]=="Sprty"])
ax[1][1].set_title('Sporty')
ax[0][1].boxplot(df["Price"][df["Type"]=="Van"])
ax[0][1].set_title('Van')
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
fig.set_figwidth(10)
fig.set_figheight(7)
ax1.boxplot(df["Horsepower"])
ax2.boxplot(df["MPG.city"])
ax1.set_title('HorsePower')
ax1.set_xlabel('HorsePowwer')
ax2.set_title('City Mileage')
ax2.set_xlabel("city milages(in miles per us gallon)")
fig.tight_layout()
```

```python
ax=df.plot(["Horsepower"],["MPG.city"],kind="scatter",color="black",marker="*",figsize=(10,7))
ax.set_xlabel("Horsepower")
ax.set_ylabel("MPG.city")
ax.set_title("Horsepower vs MPG.city",fontsize=16)                                   
```

```python
fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
plt.bar(df["DriveTrain"],df["MPG.city"],width=0.2,label="Mileage in city")
plt.suptitle("DriveTrain vs MPG.city",fontsize=16)
plt.xlabel("DriveTrain")
plt.ylabel("MPG>city")
plt.legend()                           
```
